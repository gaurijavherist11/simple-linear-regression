```python
#Train Test SPlit
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 42)
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
```

Very imp interview question:

We do not use .fit_transform for X_test because whenever we say transform in X_test data we use mean and standard deviation from X_train of data and apply the same formula is Z Score = x-mean/standard deviation because if we do the mean and standard deviation of X_test it will leak data to X_train so it wil be a problem because it will understand X_test data asldo and we don't want that so we do this and this we will also do in PCA 

```python
X_test=scaler.transform(X_test)
```

```python
## Apply Simple Linear regression
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression(n_jobs=-1) #intialize the object
```

```python
regression.fit(X_train, y_train)
```

```python
print("Coefficient or slop:",regression.coef_) #this is a slop and we got single value because 
print("Intercept:",regression.intercept_)
```

 we got single value as coefficient because this comes in y hat = beta0(intercept , we can get using regression.intercept_)+ beta1(coefficient which we calculated in upper cell). x1(x1 are data points and nothing but weights)

```python
## w.e.t training data we plot best fit line
plt.scatter(X_train, y_train)
plt.plot(X_train, regression.predict(X_train))
```

### Prediction of test data
1. Predicted salary output = intercept_ + coef_
2. y_pred_test = 70417 + 25063(X_test)

```python
##Prediction for test data
y_pred = regression.predict(X_test)
```

```python
#Performance metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

here getting big errors because we are comparing x scaled values with original data in ytest and y_pred so we will scaled y_test and y_pred also

```python
# Scale X
scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)

# Scale y
scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled  = scaler_y.transform(y_test.values.reshape(-1, 1))   # FIXED

# Train model on SCALED y
regression = LinearRegression()
regression.fit(X_train, y_train_scaled)

# Predict SCALED y
y_pred_scaled = regression.predict(X_test)

# Compare scaled values
mse = mean_squared_error(y_test_scaled, y_pred_scaled)
mae = mean_absolute_error(y_test_scaled, y_pred_scaled)
rmse = np.sqrt(mse)

print(mse)
print(mae)
print(rmse)
```

## R square
formula
**R^2 = 1 - SSR/SST**
R^2= coefficient of determination
SSR = sum of square of residual
SST = Total sum of squares


```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test, y_pred)
print(score)
```

**Adjusted R2 = 1 - [(1-R2)*(n-1)/(n-k-1)]**
where,
R2 = The R2 of the model
n = The number of the observations
k = The number of the predicted variables 

```python
#display adjusted R-squared
1 - (1 - score) * len((y_test)-1)/len((y_test)-X_test.shape[1]-1)
```
